## Ejercicios #3

Halle las ecuaciones parametricas de la recta tangente  $ f(t)=f(2cos(t),2sent(t),3t3)  $ en $t=\dfrac{\pi}{2}$

Graficando:

```python
from sympy import symbols
ax = plt.figure().add_subplot(projection='3d')
t= np.linspace(-2* np.pi, 2 * np.pi, 50)
Fx=2*np.cos(t)
Fy=2*np.sin(t)
Fz=3*t
ax.plot(Fx,Fy,Fz)

```

```python
t=symbols('t')
Fx=2*cos(t)
Fy=2*sin(t)
Fz=3*t
#Derivando
dFx=Fx.diff(t)
dFy=Fy.diff(t)
dFz=Fz.diff(t)
print('Gradiente:',dFx,' , ',dFy,' , ',dFz)
#Evaluando en t=pi 
t_O=pi
Fx_to=Fx.subs(t,t_O)
Fy_to=Fy.subs(t,t_O)
Fz_to=Fz.subs(t,t_O)
print('Punto en f(t0):',Fx_to,' , ',Fy_to,' , ',Fz_to)
dFx_to=dFx.subs(t,t_O)
dFy_to=dFy.subs(t,t_O)
dFz_to=dFz.subs(t,t_O)
print('Vector Dfto:',dFx_to,' , ',dFy_to,' , ',dFz_to)
```

Recta tangente buscada: 
$ r(t)=r(to)+t*\triangledown F()$

```python
t=symbols('t')
R_x=Fx_to+dFx_to*t
R_y=Fy_to+dFy_to*t
R_z=Fz_to+dFz_to*t
print('Ecuacion de la recta:',R_x,' , ',R_y,' , ',R_z)
```
